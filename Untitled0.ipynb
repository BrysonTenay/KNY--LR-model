```python
import pandas as pd
one_piece = pd.read_csv('/content/accidents_driver_age_data.csv')
print("dataset shape is; ", one_piece.shape)
one_piece.head()
```

```python
data = one_piece.loc[:, ['Driver Age (years)','Number of Accidents']]
data.head(5)
```

```python
import matplotlib.pyplot as plt
one_piece.plot(x='Driver Age (years)',y='Number of Accidents',style='o')
plt.xlabel('Driver Age (years)')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
x=one_piece[['Driver Age (years)']]
y=one_piece[['Number of Accidents']]
regressor = LinearRegression()
regressor.fit(x,y)
one_piece.plot(x='Driver Age (years)',y='Number of Accidents',style ='o',label='Data Points')
plt.plot(x,regressor.predict(x),color='red',label='Regression Line')
plt.xlabel('Driver Age (years)')
plt.ylabel('Number of Accidents')
plt.legend()
plt.show()
```

```python
x=one_piece[['Driver Age (years)']]
y=one_piece[['Number of Accidents']]

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)


print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

```python
print(regressor.intercept_)
print(regressor.coef_)
```

```python
y_pred = regressor.predict(x_test)
y_pred
```

```python
y_test.head(10)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
print('Mean Absolute Error:', mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
import joblib
joblib.dump(regressor, 'KNY.pkl')
```
